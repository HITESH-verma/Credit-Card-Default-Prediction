```python
bill_cols = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']
pay_amt_cols = ['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
defaulters = df[df['default'] == 1]

avg_pay = defaulters[pay_amt_cols].mean()
avg_bill = defaulters[bill_cols].mean()

plt.figure(figsize=(12, 6))

plt.plot(avg_bill.index, avg_bill.values, marker='o', label='Average Bill Amount')
plt.plot(avg_pay.index, avg_pay.values, marker='o', label='Average Payment Amount')
plt.title("Average Monthly Bill vs Payment Amount")
plt.xlabel("Month (1 = Sept, 6 = April)")
plt.ylabel("Amount (NTD)")
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), cmap='coolwarm', center=0, annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Define payment delay columns
pay_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

# Separate defaulters and non-defaulters
non_defaulters = df[df['default'] == 0]
defaulters = df[df['default'] == 1]

# Calculate mean delay for both groups
avg_delay_nondef = non_defaulters[pay_cols].mean()
avg_delay_def = defaulters[pay_cols].mean()

# Plotting
x = range(len(pay_cols))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x, avg_delay_nondef, width=width, label='Non-Defaulters', color='skyblue')
plt.bar([i + width for i in x], avg_delay_def, width=width, label='Defaulters', color='salmon')

plt.xticks([i + width/2 for i in x], pay_cols)
plt.ylabel('Average Delay (months)')
plt.title('Average Payment Delay Over 6 Months (Defaulters vs Non-Defaulters)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
